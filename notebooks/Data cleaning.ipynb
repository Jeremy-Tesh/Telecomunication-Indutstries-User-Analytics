# Handling Missing Values 

```python
#Total number of data
print(f'There are {df.shape[0]} rows and {df.shape[1]} columns')
```

```python
#Function to calculate percentage of missing values
def percent_missing(df):
    
    #Calculate total number of cells
    totalCells = np.product(df.shape)
    
    #Count number of missing values per column
    missingCount = df.isnull().sum()
    
    #Calculate total number of missing values
    totalMissing = missingCount.sum()
    
    #Calculate percentage of missing values
    print("The telecom datasets contains", round(((totalMissing/totalCells)*100),2),"%","missing values")
    
percent_missing(df)
```

```python
# Missing value amount of each column
df.isna().sum()
```

```python
# Drop columns with more than 80% missing values
limitPer = len(df) * .80
df_clean = df.dropna(thresh=limitPer, axis=1)
df_clean.isna().sum()
```

```python
features_to_not_be_null = ['Start', 'IMEI', 'IMSI','MSISDN/Number']
for feature in features_to_not_be_null:
  df_clean = df_clean[df_clean[feature].notnull()]
```

```python
df_clean.isnull().sum()
```

```python
# fill missing with ffill method for 

def fix_missing_ffill(df, col):
    df[col] = df[col].fillna(method='ffill')
    return df[col]

def fix_missing_bfill(df, col):
    df[col] = df[col].fillna(method='bfill')
    return df[col]

def fix_missing_mean(df, col):
    df[col] = df[col].fillna(df[col].mean())
    return df[col]

def fix_missing_median(df, col):
    df[col] = df[col].fillna(df[col].median())  
    return df[col]

def fix_missing_mode(df, col):
    df[col] = df[col].fillna(df_clean[col].mode()[0])
    return df[col]


```

```python
df_clean['MSISDN/Number'] = fix_missing_ffill(df_clean, 'MSISDN/Number')
df_clean['Handset Type'] = fix_missing_ffill(df_clean, 'Handset Type')
df_clean['Handset Manufacturer'] = fix_missing_ffill(df_clean, 'Handset Manufacturer')
df_clean['IMSI'] = fix_missing_ffill(df_clean, 'IMSI')


df_clean["Last Location Name"] = fix_missing_mode(df_clean,"Last Location Name")
missing_columns = ["Bearer Id",
                   "Avg RTT DL (ms)",
                   "Avg RTT UL (ms)",
                   "DL TP < 50 Kbps (%)",
                   "50 Kbps < DL TP < 250 Kbps (%)",
                   "250 Kbps < DL TP < 1 Mbps (%)",
                   "DL TP > 1 Mbps (%)",
                   "UL TP < 10 Kbps (%)",
                   "10 Kbps < UL TP < 50 Kbps (%)",
                   "50 Kbps < UL TP < 300 Kbps (%)",
                   "UL TP > 300 Kbps (%)",
                   "Nb of sec with Vol DL < 6250B",
                   "Nb of sec with Vol UL < 1250B"]

for i in missing_columns:
    df_clean[i] = fix_missing_median(df_clean, i)
```

```python
# Function to calculate missing values by column
def missing_values_table(df):
    # Total missing values
    mis_val = df.isnull().sum()

    # Percentage of missing values
    mis_val_percent = 100 * df.isnull().sum() / len(df)

    # dtype of missing values
    mis_val_dtype = df.dtypes

    # Make a table with the results
    mis_val_table = pd.concat([mis_val, mis_val_percent, mis_val_dtype], axis=1)

    # Rename the columns
    mis_val_table_ren_columns = mis_val_table.rename(
    columns = {0 : 'Missing Values', 1 : '% of Total Values', 2: 'Dtype'})

    # Sort the table by percentage of missing descending
    mis_val_table_ren_columns = mis_val_table_ren_columns[
        mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
    '% of Total Values', ascending=False).round(1)

    # Print some summary information
    print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
        "There are " + str(mis_val_table_ren_columns.shape[0]) +
          " columns that have missing values.")

    # Return the dataframe with missing information
    return mis_val_table_ren_columns

def format_float(value):
    return f'{value:,.2f}'

def find_agg(df:pd.DataFrame, agg_column:str, agg_metric:str, col_name:str, top:int, order=False )->pd.DataFrame:
    
    new_df = df.groupby(agg_column)[agg_column].agg(agg_metric).reset_index(name=col_name).\
                        sort_values(by=col_name, ascending=order)[:top]
    
    return new_df

def convert_bytes_to_megabytes(df, bytes_data):
    """
        This function takes the dataframe and the column which has the bytes values
        returns the megabytesof that value
        
        Args:
        -----
        df: dataframe
        bytes_data: column with bytes values
        
        Returns:
        --------
        A series
    """
    
    megabyte = 1*10e+5
    df[bytes_data] = df[bytes_data] / megabyte
    
    return df[bytes_data]

pd.options.display.float_format = format_float
```

```python
xDR_users = df_clean['IMSI'].value_counts()
session_duration = Users['Dur. (ms)'].sum()
Total_session_UL = Users['Total UL'].sum()
Total_session_DL = Users['Total DL'].sum()
Total_session_mb = Users['Total DL'].sum() + Users['Total UL'].sum()
Total_Social_mb = Users['Social Media DL'].sum() + Users['Social Media UL'].sum()
Total_Google_mb = Users['Google DL'].sum() + Users['Google UL'].sum()
Total_Email_mb= Users['Email DL'].sum() + Users['Email UL'].sum()
Total_Youtube_mb = Users['Youtube DL'].sum() + Users['Youtube UL'].sum()
Total_Netflix_mb = Users['Netflix DL'].sum() + Users['Netflix UL'].sum()
Total_Gaming_mb = Users['Gaming DL'].sum() + Users['Gaming UL'].sum()
Total_Other_mb= Users['Other DL'].sum() + Users['Other UL'].sum()
```

```python
sessions_df = pd.concat([xDR_users,
                 session_duration,
                 Total_session_UL,
                 Total_session_DL,
                 Total_session_mb,
                 Total_Social_mb,
                 Total_Google_mb,
                 Total_Email_mb,
                 Total_Youtube_mb,
                 Total_Netflix_mb,
                 Total_Gaming_mb,
                 Total_Other_mb], axis="columns", sort=True)
sessions_df.head()
```

```python
scaled_sessions_df = (sessions_df-sessions_df.min())/(sessions_df.max()-sessions_df.min())
scaled_sessions_df.head()
```

# Fix Outlier

```python
# function to fix outliers
def fix_outlier(df, column):
    df[column] = np.where(df[column]> df[column].quantile(0.95), df[column].median(),df[column])
    
    return df[column]



for item in scaled_sessions_df.columns:
    scaled_sessions_df[item]=fix_outlier(scaled_sessions_df, item)
    
    
scaled_sessions_df.head()
```
