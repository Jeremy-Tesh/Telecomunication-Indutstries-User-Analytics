# Non graphical univariate analysis

```python
def format_float(value):
    return f'{value:,.2f}'

def convert_bytes_to_megabytes(df, col):
    megabyte = 1000.0**2
    return df[col] / megabyte

def fix_outlier(df, column_list):
    for column in column_list:
        df[column] = np.where(df[column] > df[column].quantile(0.95), df[column].mean(),df[column])
    
    return df

pd.options.display.float_format = format_float
```

```python
data = fix_outlier(data, list(data.columns))


byte_columns = [ 'Social Media DL',
 'Social Media UL',
 'Google DL',
 'Google UL',
 'Email DL',
 'Email UL',
 'Youtube DL',
 'Youtube UL',
 'Netflix DL',
 'Netflix UL',
 'Gaming DL',
 'Gaming UL',
 'Other DL',
 'Other UL',
 'Total UL',
 'Total DL']


for col in byte_columns:
    data[col]=convert_bytes_to_megabytes(data,col)


```

```python
data.sort_values('Gaming DL', ascending=False)
```

```python
data.describe().transpose()
```

```python
def univariate_nongraphical(df):
    return df.describe()
```

```python
# univariate non graphical analysis of session duration 
univariate_dur = univariate_nongraphical(data["Dur. (ms)"])
univariate_dur
```

```python
# univariate non graphical analysis of total UL data 
univariate_totUL = univariate_nongraphical(data["Total UL"])
univariate_totDL = univariate_nongraphical(data["Total DL"])
print(univariate_totUL,"\n")
print(univariate_totUL)
```

```python
# univariate non-graphical analysis of total volume of data in megabytes for each application
# store total dl/ul data volume for each application in a dictionary
app_names = ['Social Media', 'Google', 'Email', 'Youtube', 'Netflix', 'Gaming', 'Other']
app_ul =pd.DataFrame({app_name + ' UL' : data[app_name+' UL'] for app_name in app_names})
app_dl =pd.DataFrame({app_name+' DL' : data[app_name+' DL'] for app_name in app_names})
univariate_app_ul= univariate_nongraphical(app_ul)
univariate_app_dl =univariate_nongraphical(app_dl)
print("univariate analysis for application ul data \n\n")
print(univariate_app_ul,"\n")
print("univariate analysis for application dl data \n\n")
print(univariate_app_dl)
```

#  Graphical Univariate Analysis

```python
# Different function to visualize
import matplotlib.pyplot as plt
import seaborn as sns


def plot_hist(df:pd.DataFrame, column:str, color:str)->None:
   
    sns.displot(data=df, x=column, color=color, kde=True, height=7, aspect=2)
    plt.title(f'Distribution of {column}', size=20, fontweight='bold')
    plt.show()

def plot_count(df:pd.DataFrame, column:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.countplot(data=df, x=column)
    plt.title(f'Distribution of {column}', size=20, fontweight='bold')
    plt.show()
    
def plot_bar(df:pd.DataFrame, x_col:str, y_col:str, title:str, xlabel:str, ylabel:str)->None:
    plt.figure(figsize=(12, 7))
    sns.barplot(data = df, x=x_col, y=y_col)
    plt.title(title, size=20)
    plt.xticks(rotation=75, fontsize=14)
    plt.yticks( fontsize=14)
    plt.xlabel(xlabel, fontsize=16)
    plt.ylabel(ylabel, fontsize=16)
    plt.show()

def plot_box(df:pd.DataFrame, x_col:str, title:str) -> None:
    plt.figure(figsize=(12, 7))
    sns.boxplot(data = df, x=x_col)
    plt.title(title, size=20)
    plt.xticks(rotation=75, fontsize=14)
    plt.show()
```

```python
plot_box(data, "Dur. (ms)", "Total user session duration outliers")
```

```python
plot_hist(data, "Total UL", "red")
```
